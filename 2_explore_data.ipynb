## Load and transform data

### Load libraries

```python
# conda install pandas
# conda install xlrd
# conda install openpyxl
# conda install numpy
# pip install sktime[all_extras]

# Libraries

import pandas as pd
import numpy as np
import sktime as sktime
import warnings
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sktime.utils.plotting import plot_series
from matplotlib.dates import DateFormatter
from warnings import simplefilter

# Filter warnings

warnings.filterwarnings("ignore")
```

### Load data

```python
data = pd.read_excel("Output/data_ready.xlsx")
data
```

### Explore

```python
columns = data.columns

variables = columns[columns != 'Date']

for i in variables:
    
    # Variables
    
    var = i
    
    # Time series
    
    df = data[['Date',var]]
    ts = pd.Series(df[var].values, index = df['Date'])
    
    # Plot object 
    
    fig, ax = plot_series(ts)
    
    # Dates
    
    x_dates = df['Date'].dt.strftime('%Y-%m-%d').sort_values().unique()
    ax.xaxis.set_major_formatter(plt.FixedFormatter(x_dates))
    
    # Locators
    
    ax.set_xticklabels(x_dates)
    
    # Save
    
    ax.set(title = var, xlabel='Time', ylabel='Value')
    fig.savefig('Graphics/' + var + '.png')
```
